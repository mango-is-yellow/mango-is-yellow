```python
##사업체수
import pandas as pd

f = pd.read_table("C://Users//82106//Downloads//_census_reqdoc_1659063002585//original//11_2019년_산업분류별(10차_대분류)_사업체수.txt",sep='^',header=None)
f = f.fillna(0)
f = f.iloc[:,1:]
f.iloc[:,0] = f.iloc[:,0].map(str)
f['읍면동코드'],f['일련번호'] = map(f.iloc[:,0].str.slice,[0,7],[7,12])
#print(f)

d = pd.read_excel("C://Users//82106//Downloads//ref_code//original//adm_code.xls",sheet_name='2019년6월')
d = d[d.iloc[:,0]=='11'] #서울특별시 정보만 추출
d = d.iloc[:,2:] #시도코드, 시도명칭 삭제
#print(d)

g = pd.read_excel("C://Users//82106//Downloads//ref_code//original//statistics_code.xls",sheet_name='집계구')
g = g.iloc[:,2:]
g = g[g.iloc[:,1].str.contains("cp_bnu_")] #원하는 사업체수 코드를 가진 row만 추출
g = g.tail(20)
#print(g)


#merge the dataframe
merge1 = pd.merge(f,d,left_on='읍면동코드',right_on=d.iloc[:,2],how='left')
merge1 = merge1.iloc[:,[0,2,4,6,8]]
merge2 = pd.merge(f,g,left_on=f.iloc[:,1],right_on=g.iloc[:,1],how='left')
merge2 = merge2.iloc[:,[1,6]]
sa_up_chae_soo = pd.concat([merge1,merge2],axis=1,join='inner')
sa_up_chae_soo = sa_up_chae_soo.iloc[:,[1,2,3,4,6]]
names = ['사업체수','일련번호','구이름','동이름','산업종류']
sa_up_chae_soo.columns = names
names = names[1:]+names[:1]
sa_up_chae_soo = sa_up_chae_soo[names]
print(sa_up_chae_soo)


##grouping by Dong name
#sucs = sa_up_chae_soo.groupby('동이름')
#sucs.get_group('가락1동')
```
